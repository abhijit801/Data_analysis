```python
ipl_del['match_id'] = ipl_del['match_id'].astype('int32') # reducing unnecessary memory size...
ipl_del['inning'] = ipl_del['inning'].astype('int32')
ipl_del['over'] = ipl_del['over'].astype('float32')
ipl_del['ball'] = ipl_del['ball'].astype('float32')
ipl_del['is_super_over'] = ipl_del['is_super_over'].astype('int32')
ipl_del['wide_runs'] = ipl_del['wide_runs'].astype('int32')
ipl_del['bye_runs'] = ipl_del['bye_runs'].astype('int32')
ipl_del['legbye_runs'] = ipl_del['legbye_runs'].astype('int32')
ipl_del['noball_runs'] = ipl_del['noball_runs'].astype('int32')
ipl_del['penalty_runs'] = ipl_del['penalty_runs'].astype('int32')
ipl_del['batsman_runs'] = ipl_del['batsman_runs'].astype('int32')
ipl_del['extra_runs'] = ipl_del['extra_runs'].astype('int32')
ipl_del['total_runs'] = ipl_del['total_runs'].astype('int32')
```

```python
ipl.drop(columns=['method'],inplace=True) # droping unnecssary column...
```

```python
ipl_del.info()
```

```python
# Every season final winner team...
final_winner = ipl[ipl['MatchNumber']=="Final"][['Season','WinningTeam']]   
final_winner = final_winner.reset_index(drop=True)
final_winner
```

```python
# how many times have  super overs been played in ipl...
super_over = ipl[ipl['SuperOver']=='Y'].shape[0]
super_over
```

```python
# ipl final tosswinner team percentage to match win.....
final = ipl[(ipl['MatchNumber']=='Final')].shape[0]
percentage = ipl[(ipl['MatchNumber'] == 'Final') & (ipl['TossWinner'] == ipl['WinningTeam'])].shape[0]/final*100 
percentage
```

```python
# In playoff matches most player of match win players...
ipl[~ipl['MatchNumber'].str.isdigit()]['Player_of_Match'].value_counts().head(10)
```

```python
# Decision taken by the toss winnning team...
ipl['TossDecision'].value_counts()
```

```python
(ipl['Team1'].value_counts() + ipl['Team2'].value_counts()).sort_values(ascending = False)
```

```python
# cak win in kolkata...
ipl[(ipl['City'] == 'Kolkata') & (ipl['WinningTeam'] == 'Chennai Super Kings')].shape[0]
```

```python
# top 10 run scorer in ipl...
ipl_del.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
# top 10 six hited batsmans .....
six = ipl_del[ipl_del['batsman_runs'] == 6].groupby('batsman')['batsman'].count().sort_values(ascending = False).head(10)
six
```

```python
# here top 10 batsman hit most baundries in last 5 overs...
last = ipl_del[ipl_del['over']>15]
last = last[(last['batsman_runs'] == 4) | (last['batsman_runs']==6)]
last.groupby('batsman')['batsman'].count().sort_values(ascending = False).head(10)
```

```python
vk = ipl_del[ipl_del['batsman'] == 'V Kohli'] # virat every team against runs...
vk = vk.groupby('bowling_team')['batsman_runs'].sum().reset_index().sort_values(by='batsman_runs',ascending = False)
vk
```

```python
# rcb highest wicket taker...
wicket_taker = ipl_del[ipl_del['bowling_team'] == 'Royal Challengers Bangalore'].groupby('bowler')['player_dismissed'].count().sort_values(ascending = False).head(10)
wicket_taker
```
